# Seaborn 

Seaborn matplotlib gibi popüler bir görselleştirme kütüphanesi.



```python
import seaborn as sns
```

```python
sns.__version__
```

```python
sns.get_dataset_names() # bunlar hazır olarak seabornda bulunan veri setlerini gösterir.
```

```python
# tips veri setini kullanalım.
tips=sns.load_dataset('tips')
tips.head()
```

```python
# seborn ile bu veri setinden basit bir bar grafiği çizersek.
# x ekseni ve y eksenini belirleyebiliriz.
sns.barplot(x='day',y='total_bill',data=tips)
```

```python
# seaborn kütüphanesinin en önemli özelliklerinden olan hue parametresi;
sns.barplot(x='day',y='total_bill',hue='sex',data=tips)
```

```python
sns.barplot(x='day',y='total_bill',hue='smoker',data=tips)
# bu özellikle her çubuk sigara içen veya içmeyen olarak 2 ye ayrılıyor.
```

```python
#Seaborn kütüphanesinin bir çok renk paleti bulunur.
sns.color_palette() # gösterilen renkler deep 

# deep, ligt, winter_r, spring
```

```python
sns.color_palette('spring',10) #spring panelinden 10 renk
```

```python
sns.color_palette('winter_r',10)
```

```python
# Seborn da özel olarak paket oluşturulabilir.
new_palette=['#8A2BE2','#4B0082','#483D8B','#6A5ACD','#8B008B'] 
# kullanılacak renk kodları veya adıyla bir dizi oluşturulur.
```

```python
sns.set_palette(new_palette)
sns.color_palette() #listedeki renklerle yeni bir palet oluşturulmus oldu.
```

```python
sns.barplot(x='total_bill',y='day',data=tips) 
# x ve y ekseni değerleri yer değiştirildiğinde grafik yatay bir şekil alır
```

```python
# grafikteki sıralama da değiştirilebilir.
sns.barplot(x='day',y='total_bill',order=['Fri','Thur','Sat','Sun'],data=tips)
```

```python
# grafiği kaydetmek
sns_plot= sns.barplot(x='day',y='total_bill',order=['Fri','Thur','Sat','Sun'],data=tips)
```

```python
fig= sns_plot.get_figure()
```

```python
fig.savefig(fname='plot.png')
```

### DistPlots 

```python
penguins=sns.load_dataset('penguins')
penguins.head()
```

```python
sns.distplot(penguins.flipper_length_mm)
# flipper_length_mm palet uzunluğu 
```

```python
# rug özelliği, tabanda işaretleme yapar.
sns.distplot(penguins.flipper_length_mm, rug=True) 

# bu rug özellikleri değiştirilebilir.
sns.distplot(penguins.flipper_length_mm, rug_kws={'color':'blue','height':0.1},rug=True) 
```

```python
# grafikteki çizgiyi (KDE) veya histogram kaldırılabilir.

sns.distplot(penguins.flipper_length_mm, kde=False)
sns.distplot(penguins.flipper_length_mm, hist=False)
```

```python
# bin 
# kutuların boyutunu ayarlar önemli bir parametredir. 
# Kutuların yanlış boyutu alması verilerin önemli özelliklerini gizleyebilir.
# kutu boyutu binwidth parametresiyle 
# kutu sayısı bins parametresiyle ayarlanır.
# bazı durumlarda boyutlardan ziyade kutu sayısını belirlemek daha mantıklı olabiliyor.

sns.distplot(penguins.flipper_length_mm, bins=20)
```

```python
# hue
# verilerde benzersiz değerlerin her biri için ayrı histogram çizer ver renklere ayırır.
sns.displot(penguins, x="flipper_length_mm", hue="species")
```

```python
# Varsayılan olarak, farklı histogramlar üst üste "katmanlıdır" ve bazı durumlarda ayırt edilmeleri zor olabilir. 
# Seçeneklerden biri, histogramın görsel sunumunu bir çubuk grafikten bir "step" grafiğine değiştirmektir:
sns.displot(penguins, x="flipper_length_mm", hue="species", element="step")
```

```python
# her bir ayrı dağılımı ayrı bir alt grafikte çizmek de mümkündür.
# örneğin cinsiyete göre bir değerlendirme yapıldığında 
sns.displot(penguins, x="flipper_length_mm", col="sex")
# col -> yan yana
# row -> alt alta 
# hue -> bir grafikte üst üste gösterir. 
```

## Normalleştirilmiş histogram
Bunu yapmadan önce, dikkat edilmesi gereken bir diğer nokta da, alt kümeler eşit olmayan sayıda gözleme sahip olduğunda, dağılımlarını sayım cinsinden karşılaştırmak ideal olmayabilir. Bir çözüm, parametreyi kullanarak sayıları normalleştirmektir 

```python
sns.displot(penguins, x="flipper_length_mm", hue="species", stat="density")
```

```python
# Bununla birlikte, varsayılan olarak, normalleştirme tüm dağılıma uygulanır, 
# bu nedenle bu sadece çubukların yüksekliğini yeniden ölçeklendirir. ayarlandığında
# common_norm=False, 
# her bir alt küme bağımsız olarak normalleştirilir:
sns.displot(penguins, x="flipper_length_mm", hue="species", stat="density", common_norm=False)
```

```python
# Bazı durumlarda histogram yerine çekirdek yoğunluğu kullanılması daha doğru olabilir.
# kind-> kde 

sns.displot(penguins, x="flipper_length_mm", kind="kde")
```

```python
# Kde nin verileri doğru temsil etmesi bant genişliğine bağlıdır.
# aşırı düzleştirilmiş bir tahmin anlamlı özellikleri silebilir.
# yeterince yumuşatılmamıs bir tahmin gerçek şekli rastgele gürültüye gizleyebilir.

sns.displot(penguins, x="flipper_length_mm", kind="kde", bw_adjust=.25)
```

```python
sns.displot(penguins, x="flipper_length_mm", kind="kde", bw_adjust=2)
```

```python
# histogramdaki gibi hue kullanılarak değişkenin her düzeyi için ayrı renk kullanır.
sns.displot(penguins, x="flipper_length_mm", hue="species", kind="kde")
```

```python
# bu grafiklerin içi doldurulabilir. 
sns.displot(penguins, x="flipper_length_mm", hue="species", kind="kde",
            multiple="stack")
```

```python
# dağılımın arka planda nasıl olduğunu görmek için opaklık ayarlanabilir. 
# Böylece her birinin hareketi daha net gözlemlenir.
sns.displot(penguins, x="flipper_length_mm", hue="species", kind="kde", fill=True)
```
