## Axes labels, Legend, Grid

```python
days=[1,2,3,4,5,6,7]
max_t=[50,51,52,48,47,49,46]
min_t=[43,42,40,44,33,35,37]
avg_t=[45,48,48,46,40,42,41]
```

```python
# btn deerler ayn grafikte gsterilir.
plt.xlabel=("days")
plt.ylabel=("tamperature")
plt.title=("weather")
# her bir izginin hangi deerlere ait olduu etiketlenebilir. 
plt.plot(days,max_t, label="Max")
plt.plot(days,min_t, label="Min")
plt.plot(days,avg_t, label="Avg")

plt.legend() # etiketlerin gsterilmesini salar
# Bu etiketler belirli yerlerde konumlanabilir.
# rn
# plt.legend(loc="lower right")
# plt.legend(loc="best") bu da modul en iyi nerede olacana kendisi karar verir.

plt.legend(shadow=True, fontsize="large") # gibi zelliklerle de bu gsterim netletirilebilir.

plt.grid() # grafii zgara eklinde gstererek noktalar iin bir grsel salar.
```

```python
company= ["GOOGLE","AMZN","MSFT","FB"]
revenue=[90,136,89,27]
profit=[40,2,34,12]
```

```python
# kazanlarla beraber kar deerlerini de grmek iin ayn bar grafii kullanlr

plt.xticks(xpos,company)
plt.bar(xpos,revenue, label="Revenue")
plt.bar(xpos,profit, label="Profit")
plt.legend()

# Bu kullanmda ayn barda gsterilir. Ancak yan yana barda gsterilmesi daha mantkldr.
```

```python
# yan yana barlarda bu deerler gsterilmek istendiinde  
plt.xticks(xpos,company)
plt.bar(xpos- 0.2, revenue, label="Revenue")
plt.bar(xpos + 0.2, profit, label="Profit")
plt.legend()

# burada +0.2 ve -0.2 mant 1-0.2 yani 0.8 gibi bir konumu olacak
# yani Revenue 1 den 0.2 aa bir konumda
# Profit 1 den 0.2 yukar bir konumda balayacak ve bylece barlar birbirinden 
# ayr grnecektir.
```

```python
# tamamen yan yana iki grafik istendiinde geniliklere mdahale edilir. 
plt.xticks(xpos,company)
plt.bar(xpos- 0.2, revenue, width=0.4,label="Revenue")
plt.bar(xpos + 0.2, profit, width=0.4,label="Profit")
plt.legend()

# genilikler telenme deerlerine gre ayarlandnda yan yana barlar elde edilir.
# r xpos-0.3 olsayd width 0.6 olarak byklkleri farkl ama yan yana bar elde edilir.
```

```python
# grafik dikey deil de yatay olarak izilmek istenilirse
# barh kullanlr.
# btn ifadeler de yatay dzeleme gre yeniden dzenlenir.

plt.yticks(xpos,company)
plt.barh(xpos- 0.2, revenue,height=0.4, label="Revenue")
plt.barh(xpos + 0.2, profit, height=0.4, label="Profit")
plt.legend()
```

## Histograms

```python
# histogram tek boyutlu diziyle alsr.
blood_sugar=[113,85,90,150,149,88,93,115,135,80,77,82,129]
```

```python
plt.hist(blood_sugar)
```

```python
# yukardaki histogramda default olarak 10 blme izer. 
# Bu blmeler deerlere gre ayarlanabilir.

# r 3 aralk isteniyorsa bins=3
# plt.hist(blood_sugar, bins=3,rwidth=0.85)

plt.hist(blood_sugar, bins=[80,100,125,160],rwidth=0.85, color="purple",)
# burada histogramn aralklar (seviyeler) dizi olarak belirlenmi

# rwidth barlarn geniliiyle alakaldr. 

plt.hist(blood_sugar, bins=[80,100,125,160],rwidth=0.85, color="purple")
```

```python
# iki cinsiyet iin byle veriler olduu durumda;
blood_sugar_men=[113,85,90,150,149,88,93,115,135,80,77,82,129]
blood_sugar_women=[67,98,89,120,133,150,84,69,89,79,120,112,100]
```

```python
# bu durumda histogram deerleri bir dizi olarak gnderilir. 

plt.hist([blood_sugar_men, blood_sugar_women], bins=[80,100,125,160],rwidth=0.85, color=["purple","blue"], 
         label=["men","women"], orientation='vertical')
# orientation='horizontal' yatay olarak gstermek istenirse 
plt.legend()
```

## Pie Chart

```python
exp_vals=[1400,600,300,410,250]
exp_labels=["home","food","phone/internet","car","other"]
```

```python
plt.pie(exp_vals, labels= exp_labels)
plt.show() # fazladan olan etiketleri-deerleri gstermez

# grafiin boyutunu deitirmek iin yarap- radius zellii kullanlr.
# default deeri 1 
plt.pie(exp_vals, labels= exp_labels, radius=1.5)
plt.show() 
```

```python
# bu deerleri daha ie yarar bir ekilde gstermek iin 
# bu giderlerin yzde ka olduu gsterilebilir.

plt.pie(exp_vals, labels= exp_labels, radius=1.2, autopct= '%0.1f%%', shadow=True)
# autopct= '%f%%'-> yzdelik olarak bir gsterimdir. 
# %0.2f%% -> virglden sonra 2 basamak gsterir.
#  %0.0f%% -> virgll ksm yoktur.

plt.show() 
```

```python
# bu dilimler karlmak istenirse;
plt.pie(exp_vals, labels= exp_labels, radius=1.2, autopct= '%0.1f%%',
        explode=[0,0.1,0,0.2,0])
# explode=[0,0,0,0,0] ekilde yani verilen deerler dizi sralamasyla 0 olarak konumlanm
# burada karlmak istenen para indexinin deeri deitirilerek o lde uzaklatrr.
plt.show() 


# buradaki sralama as deitirilmek istenirse startangle parametresi kullanlr.

plt.pie(exp_vals, labels= exp_labels, radius=1.2, autopct= '%0.1f%%',
        explode=[0,0.1,0,0.2,0], startangle=90)
# startange=90 olduunda ilk eleman 90 derecelik bir ayla balar 


plt.show() 
```

### Grafii PNG dosyas olarak kaydetmek.

```python
plt.pie(exp_vals, labels= exp_labels, radius=1.2, autopct= '%0.1f%%',
        explode=[0,0.1,0,0.1,0], startangle=90)

# plt.savefig("piechart.png") bu fonksiyon png olarak kaydetmede kullanlabilir.

# ancak bu fonksiyon direkt kullanldnda resimde krplmalar oluabilir.
# bbox_inches="tight" zellii kullanldnda grafik tamamen sdrlarak kaydedilir.

# bu sdrma dsnda da ekstra alanlar braklarak kaydedilmek istenirse pad_inches kullanlr.
# verilen deer kadar boluk (in) brakarak kaydeder.

# Grafiin arka taraf transparan olsun isteniyorsa transparent=True zellii kullanlr.

plt.savefig("piechart.png", bbox_inches="tight", pad_inches=2)

# bulunan dizinden farkl bir konuma pdf olarak kaydedilmek istenirse isim banda bu yol belirtilir.
# ve png yerine pdf olarak belirtilir.
# "c:/klasor/python/piechart.pdf"
```
