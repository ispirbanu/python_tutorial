# Matplotlib
İnsan beyni görüntüleri metinden daha iyi işleme gücüne sahiptir. Bu yüzden görselleştirmeler her alanda olduğu gibi burada da önemli bir kavram.

Örneğin sıcaklıkların olduğu bir veri tablosu olsun. Bu veri tablolarında noktaları tek tek tarayarak ve değerler karşılaştırılarak en yüksek ve en düşük sıcaklıklar bulunur. 



Ancak bu noktalar görselleştirildiğinde bu değerleri tek tek karşılaştırmamız gerekmeyecek. Çünkü görür görmez beyin bunu algılayabiliyor.





```python
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
x=[1,2,3,4,5,6,7] # Günler 
y=[50,51,52,48,47,49,46] # sıcaklık
```

```python
plt.plot(x,y) # 2 boyutlu grafiği basitçe grafiği çizdirirsek 
```

```python
# plot fonksiyonu bir çok parametre alıyor. 
# color: grafik rengini belirler.
# linewidth: çizgi kalınlığını belirler.
# linestyle: çizginin şeklini belirler. (solid,dashed,dashdot,dotted)
```

```python
plt.xlabel('Day')
plt.ylabel('temperature')
plt.title=('Weather')
plt.plot(x,y,color="blue",linewidth=2, linestyle="dotted")
```

```python
# Grafik değiştirilebilir
plt.plot(x,y,'r*') # buradaki r red * ise grafikte kullanılacak şekil olarak da 
# kısaca grafik şekli belirlenebilir.
```

```python
plt.plot(x,y,'r*-.') # ifadesiyle noktalar dısında grafik şekli de görülebilir.
# başta verilen ifade grafiğin çizileceği şekildir.
```

```python
# noktaları klavyede olmayan şekillerde de gösterilebilir bunlar matplotlib sayfasında bulunuyor.
#örneğin;
plt.plot(x,y,'bD-')
# D-> diamond 
# bu ifade 
# plt.plot(x,y,color="blue", marker='D', linestyle="solid") 
# kısaltmasıdır. 
```

```python
# Grafik çzigileri saydamlaştırıladabilir.
# renkler rgb kodlarıyla da yazılabilir.

plt.plot(x,y,color="#888840", marker='D', linestyle="solid", alpha=0.5) 

# alpha=1 -> tamamen opak
# alpha=0 -> tamamen saydam
```

## Axes labels, Legend, Grid

```python
days=[1,2,3,4,5,6,7]
max_t=[50,51,52,48,47,49,46]
min_t=[43,42,40,44,33,35,37]
avg_t=[45,48,48,46,40,42,41]
```

```python
# bütün değerler aynı grafikte gösterilir.
plt.xlabel=("days")
plt.ylabel=("tamperature")
plt.title=("weather")
# her bir çizginin hangi değerlere ait olduğu etiketlenebilir. 
plt.plot(days,max_t, label="Max")
plt.plot(days,min_t, label="Min")
plt.plot(days,avg_t, label="Avg")

plt.legend() # etiketlerin gösterilmesini sağlar
# Bu etiketler belirli yerlerde konumlanabilir.
# örn
# plt.legend(loc="lower right")
# plt.legend(loc="best") bu da modul en iyi nerede olacağına kendisi karar verir.

plt.legend(shadow=True, fontsize="large") # gibi özelliklerle de bu gösterim netleştirilebilir.

plt.grid() # grafiği ızgara şeklinde göstererek noktalar için bir görsel sağlar.
```

```python
company= ["GOOGLE","AMZN","MSFT","FB"]
revenue=[90,136,89,27]
profit=[40,2,34,12]
```

```python
# kazançlarla beraber kar değerlerini de görmek için aynı bar grafiği kullanılır

plt.xticks(xpos,company)
plt.bar(xpos,revenue, label="Revenue")
plt.bar(xpos,profit, label="Profit")
plt.legend()

# Bu kullanımda aynı barda gösterilir. Ancak yan yana barda gösterilmesi daha mantıklıdır.
```

```python
# yan yana barlarda bu değerler gösterilmek istendiğinde  
plt.xticks(xpos,company)
plt.bar(xpos- 0.2, revenue, label="Revenue")
plt.bar(xpos + 0.2, profit, label="Profit")
plt.legend()

# burada +0.2 ve -0.2 mantığı 1-0.2 yani 0.8 gibi bir konumu olacak
# yani Revenue 1 den 0.2 aşağı bir konumda
# Profit 1 den 0.2 yukarı bir konumda başlayacak ve böylece barlar birbirinden 
# ayrı görünecektir.
```

```python
# tamamen yan yana iki grafik istendiğinde genişliklere müdahale edilir. 
plt.xticks(xpos,company)
plt.bar(xpos- 0.2, revenue, width=0.4,label="Revenue")
plt.bar(xpos + 0.2, profit, width=0.4,label="Profit")
plt.legend()

# genişlikler ötelenme değerlerine göre ayarlandığında yan yana barlar elde edilir.
# ör xpos-0.3 olsaydı width 0.6 olarak büyüklükleri farklı ama yan yana bar elde edilir.
```

```python
# grafik dikey değil de yatay olarak çizilmek istenilirse
# barh kullanılır.
# bütün ifadeler de yatay düzeleme göre yeniden düzenlenir.

plt.yticks(xpos,company)
plt.barh(xpos- 0.2, revenue,height=0.4, label="Revenue")
plt.barh(xpos + 0.2, profit, height=0.4, label="Profit")
plt.legend()
```

## Histograms

```python
# histogram tek boyutlu diziyle çalısır.
blood_sugar=[113,85,90,150,149,88,93,115,135,80,77,82,129]
```

```python
plt.hist(blood_sugar)
```

```python
# yukarıdaki histogramda default olarak 10 bölme çizer. 
# Bu bölmeler değerlere göre ayarlanabilir.

# ör 3 aralık isteniyorsa bins=3
# plt.hist(blood_sugar, bins=3,rwidth=0.85)

plt.hist(blood_sugar, bins=[80,100,125,160],rwidth=0.85, color="purple",)
# burada histogramın aralıkları (seviyeler) dizi olarak belirlenmiş

# rwidth barların genişliğiyle alakalıdır. 

plt.hist(blood_sugar, bins=[80,100,125,160],rwidth=0.85, color="purple")
```

```python
# iki cinsiyet için böyle veriler olduğu durumda;
blood_sugar_men=[113,85,90,150,149,88,93,115,135,80,77,82,129]
blood_sugar_women=[67,98,89,120,133,150,84,69,89,79,120,112,100]
```

```python
# bu durumda histogram değerleri bir dizi olarak gönderilir. 

plt.hist([blood_sugar_men, blood_sugar_women], bins=[80,100,125,160],rwidth=0.85, color=["purple","blue"], 
         label=["men","women"], orientation='vertical')
# orientation='horizontal' yatay olarak göstermek istenirse 
plt.legend()
```

## Pie Chart

```python
exp_vals=[1400,600,300,410,250]
exp_labels=["home","food","phone/internet","car","other"]
```

```python
plt.pie(exp_vals, labels= exp_labels)
plt.show() # fazladan olan etiketleri-değerleri göstermez

# grafiğin boyutunu değiştirmek için yarıçap- radius özelliği kullanılır.
# default değeri 1 
plt.pie(exp_vals, labels= exp_labels, radius=1.5)
plt.show() 
```

```python
# bu değerleri daha işe yarar bir şekilde göstermek için 
# bu giderlerin yüzde kaç olduğu gösterilebilir.

plt.pie(exp_vals, labels= exp_labels, radius=1.2, autopct= '%0.1f%%', shadow=True)
# autopct= '%f%%'-> yüzdelik olarak bir gösterimdir. 
# %0.2f%% -> virgülden sonra 2 basamak gösterir.
#  %0.0f%% -> virgüllü kısım yoktur.

plt.show() 
```

```python
# bu dilimler çıkarılmak istenirse;
plt.pie(exp_vals, labels= exp_labels, radius=1.2, autopct= '%0.1f%%',
        explode=[0,0.1,0,0.2,0])
# explode=[0,0,0,0,0] şekilde yani verilen değerler dizi sıralamasıyla 0 olarak konumlanmış
# burada çıkarılmak istenen parça indexinin değeri değiştirilerek o ölçüde uzaklaştırır.
plt.show() 


# buradaki sıralama açısı değiştirilmek istenirse startangle parametresi kullanılır.

plt.pie(exp_vals, labels= exp_labels, radius=1.2, autopct= '%0.1f%%',
        explode=[0,0.1,0,0.2,0], startangle=90)
# startange=90 olduğunda ilk eleman 90 derecelik bir açıyla başlar 


plt.show() 
```

### Grafiği PNG dosyası olarak kaydetmek.

```python
plt.pie(exp_vals, labels= exp_labels, radius=1.2, autopct= '%0.1f%%',
        explode=[0,0.1,0,0.1,0], startangle=90)

# plt.savefig("piechart.png") bu fonksiyon png olarak kaydetmede kullanılabilir.

# ancak bu fonksiyon direkt kullanıldığında resimde kırpılmalar oluşabilir.
# bbox_inches="tight" özelliği kullanıldığında grafik tamamen sığdırılarak kaydedilir.

# bu sığdırma dısında da ekstra alanlar bırakılarak kaydedilmek istenirse pad_inches kullanılır.
# verilen değer kadar boşluk (inç) bırakarak kaydeder.

# Grafiğin arka tarafı transparan olsun isteniyorsa transparent=True özelliği kullanılır.

plt.savefig("piechart.png", bbox_inches="tight", pad_inches=2)

# bulunan dizinden farklı bir konuma pdf olarak kaydedilmek istenirse isim başında bu yol belirtilir.
# ve png yerine pdf olarak belirtilir.
# "c:/klasor/python/piechart.pdf"
```
