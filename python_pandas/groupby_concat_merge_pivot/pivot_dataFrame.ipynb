pivot DataFrame'i yeniden şekillendirmeye, döndürmeye yarar.
Örneğin sahip olunan hava durumu DataFrame'i ;


verilerin x ekseni tarih y ekseni şehirler olsun şeklinde düzenlenmek istenmesi



veya verilerin x ekseni şehir y ekseni tarihler olsun şeklinde düzenlenmek istenmesi.

Bunların hepsi pivot ile yapılabilir.

```python
import pandas as pd
```

```python
df= pd.read_csv("weather_data_pivot.csv")
df
```

```python
df.pivot(index="date", columns="city")
```

```python
# bu çevrilmiş tabloda sadece sıcaklık değerlerini görmek için;
df.pivot(index="date", columns="city", values='temperature')
```

```python
df2= pd.read_csv("weather_data_pivot.csv")
df2.pivot_table(index="city",columns="date",margins=True) 
# margins sayıların satırda ve sütunda ortalamasını gösterir
```
