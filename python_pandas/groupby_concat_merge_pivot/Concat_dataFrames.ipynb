```python
import pandas as pd
```

```python
us_weather=pd.DataFrame({
    'city': ['New York', 'Boston','San Diego'],
    'temperature':[21,14,35],
    'humidity':[68,65,75]
})
us_weather
```

```python
in_weather=pd.DataFrame({
    'city': ['mumbai', 'delhi','banglore'],
    'temperature':[31,33,43],
    'humidity':[80,60,78]
})
in_weather
```

```python
# bu iki veri setini birleştirmek için concat kullanılır.
df=pd.concat([us_weather,in_weather])
df
```

veri setinde numaralandırma birleştirilmeden önceki haliyle alınıyor. 
Bunu sağlayan concat metodunun parametresinden varsayılanı False olan "ignore_index" özelliği.
True yapıldığında indexleme 0 dan n-1 e göre sıralanır.

```python
df=pd.concat([us_weather,in_weather],ignore_index=True) 
df
```

```python
# Bu verileri amerika ve hindistan olarak birleşmesine rağmen göstermek için
# keys parametresi kullanılır.
# ! bu durumda ignore_index çalışmaz.
df=pd.concat([us_weather,in_weather],keys=["us","india"]) 
df
```

```python
# Bu gösterimin avantajı kullanılan bu key özelliğine göre gruplanmıs gibi olması
df.loc["us"]
```

Event diye bir veri setimiz daha olsun. Bu veri setini hava durumu veri setine sütun olarak eklemek için concat metodunda axis parametresi kullanılır.

```python
us_weather=pd.DataFrame({
    'city': ['New York', 'Boston','San Diego'],
    'temperature':[21,14,35],
    'humidity':[68,65,75]
})
us_weather
```

```python
event_df=pd.DataFrame({
    'city': ['New York', 'Boston','San Diego'],
    'event':['Rain','Rain','Sunny']
})
event_df
```

```python
df2= pd.concat([us_weather,event_df], axis=1)
df2
```

Burada iki verisetinde de şehirin doğru sıralanmış olması önemli aksi halde bu veriler yanlış yerleştirilir böyle bir durumda index özelliği kullanılmalı ve şehirlerin hangi indexte olduğu belirtilmeli. 

```python
us_weather=pd.DataFrame({
    'city': ['New York', 'Boston','San Diego'],
    'temperature':[21,14,35],
    'humidity':[68,65,75]
}, index=[0,1,2])

event_df=pd.DataFrame({
    'city': ['San Diego', 'Boston','New York'],
    'event':['Rain','Rain','Sunny']
}, index=[2,1,0])

df3= pd.concat([us_weather,event_df], axis=1)
df3
```
