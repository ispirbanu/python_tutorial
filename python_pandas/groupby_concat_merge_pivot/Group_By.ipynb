```python
import pandas as pd
```

```python
df= pd.read_csv("weather_data_Group_by.csv")
df
```

Bu Dataframe şehirlere göre gruplandırılırsa; 


```python
g= df.groupby('city') # şehirleri gruplayarak değişkene atıyoruz.
# Bu gruplama SQL groupby özelliğine benziyor.
# SELECT * FROM city_data GROUP BY city 
g # bir obje olarak döner.
```

```python
for c, c_df in g:
    print(c)
    print(c_df)
```

```python
# eğer gruplama dataFrame olarak görmek istenirse
g.get_group('new york')
```

```python
g.max()  # her şehirden maksimum sıcaklığa sahip verileri gösterir.
```

```python
g.describe() # şehir gruplarının istatistik özelliklerini gösterir.
```

```python
# Bu grupları görselleştirirsek
# sayısal olan sıcaklık ve rüzgar hızına göre her şehirin grafiğini verir.
%matplotlib inline
g.plot()
```
