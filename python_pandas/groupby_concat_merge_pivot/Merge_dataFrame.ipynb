```python
import pandas as pd
```

```python
us_weather=pd.DataFrame({
    'city': ['New York', 'Boston','San Diego'],
    'temperature':[21,14,35],
})
us_weather
```

```python
in_weather=pd.DataFrame({
    'city': ['San Diego','New York','Boston'],
    'humidity':[65,68,75],
})
in_weather
```

```python
# Bu iki veri setini şehirlere göre birleştirmek için
new_df= pd.merge(us_weather,in_weather, on="city")
new_df
```

verilerin sıralaması farklı olsa da birleştirmeyi doğru bir şekilde yapar. verilerde farklılıklar olsa ve veri seti boyutları uyuşmasa da ortak olan veriler için doğru birleştirmeyi yapar.
Bu inner join birleştirme şeklidir.


```python
df1=pd.DataFrame({
    'city': ['New York', 'baltimore','chicago', "Orlando"],
    'temperature':[21,14,35,32],
})

df2=pd.DataFrame({
    'city': ['San francisco','New York','chicago'],
    'humidity':[65,68,75],
})

df3=pd.merge(df1,df2,on="city")
df3
```

Diğer birleştirme şekli olan outer join ile verilerin hepsinikapsayacak birleştirme yapar. 
bunun için "how" parametresi kullanılır. Varsayılanı inner birleştirmedir. (yukarıdaki örnek) 


.




```python
df4=pd.merge(df1,df2,on="city",how="outer")
df4
```

```python
# 'how' parametresinin 'left' ve 'right' olarak sola ve sağa dayalı birleştirme şekilleri de vardır.
df5=pd.merge(df1,df2,on="city",how="right")
df5
```
